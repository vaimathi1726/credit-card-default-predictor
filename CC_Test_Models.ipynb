```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1) Read the aggregated results
df = pd.read_csv("testresults.csv")

# 2) Barplot of F1 by model
plt.figure(figsize=(6,4))
sns.barplot(
    data=df,
    x='model',
    y='f1_score',
    palette='deep'
)
plt.title('Comparison of F1 Score Across Models')
plt.ylabel('F1 Score')
plt.xlabel('Model')
plt.tight_layout()
plt.show()

# 3) (Optional) plot other metrics side by side
melted = df.melt(
    id_vars='model',
    value_vars=['accuracy','precision','recall','f1_score'],
    var_name='metric',
    value_name='score'
)
plt.figure(figsize=(8,5))
sns.barplot(
    data=melted,
    x='model',
    y='score',
    hue='metric',
    palette='Set2'
)
plt.title('All Metrics by Model')
plt.ylabel('Score')
plt.xlabel('Model')
plt.legend(title='Metric', bbox_to_anchor=(1.05,1), loc='upper left')
plt.tight_layout()
plt.show()
```
